## Question 3.2

- Create a Table of all the distinct players with a first name of Bob who were born in the United States and played at Stanford university


- Include their ```first name, last name, playerID```, and ```birth state```


- Hint: Use a Join between ```People``` and ```CollegePlaying```


- You need to return 10 rows.

<u>Answer:</u>

````python
%sql DROP TABLE IF EXISTS BOB;
````

````python
%sql CREATE TABLE BOB \
    AS SELECT DISTINCT People.nameFirst, People.nameLast, People.playerID, People.birthState, CollegePlaying.schoolID \
    FROM People \
    INNER JOIN CollegePlaying ON People.playerID=CollegePlaying.playerID \
    WHERE People.nameFirst='Bob' AND CollegePlaying.schoolID='stanford';
    
````

````python
%sql SELECT * FROM BOB LIMIT 10;
````

## Question 3.3

- Group together players with the same birth year, and report the year, the number of players in the year, and average height for the year


- Order the resulting by year in descending order. Put this in a new table called ```AverageHeight```


- Hint: height will be NULL for some of these years, do not consider these entries


- Run the SELECT * statement to show that your table was created successfully.

<u>Answer:</u>

````python
%sql DROP TABLE IF EXISTS AverageHeight;
````

````python
%sql CREATE TABLE AverageHeight AS \
    SELECT AVG(height) AS AverageHeight, birthYear \
    FROM People \
    GROUP BY birthYear \
    HAVING AverageHeight AND birthYear \
    ORDER BY birthYear desc;
````

````python
%sql SELECT * FROM AverageHeight LIMIT 10;
````

## Question 3.4

- Find the players who made it into the hall of fame who played for a college located in NJ


- Return the ```player ID, first name, last name```, and ```school ID```. 


- Order the players by ```School``` alphabetically.


- You need to return 10 rows.

<u>Answer:</u>

````python
%sql SELECT DISTINCT HallOfFame.playerID, People.nameFirst, People.nameLast, CollegePlaying.schoolID, People.birthState \
    FROM HallOfFame \
    INNER JOIN CollegePlaying ON CollegePlaying.playerID=HallOfFame.playerID \
    INNER JOIN People ON People.playerID=HallOfFame.playerID \
    WHERE People.birthState='NJ' \
    ORDER BY CollegePlaying.schoolID asc \
    LIMIT 10;
````

## Question 3.5

- Find the ```team id, yearid```, and ```average HBP``` for each team using a subquery.


- Group the entries by ```team``` and ```year``` and order by descending values


- Hint: be careful to only include entries where ```AB``` is > 0


- You need to return 50 rows.

<u>Answer:</u>

````python
%sql SELECT DISTINCT Salaries.teamID, Salaries.yearID, (SELECT AVG(Batting.HBP) \
        FROM Batting \
        INNER JOIN Salaries ON Salaries.playerID=Batting.playerID \
        WHERE Batting.AB > 0 \
        GROUP BY Salaries.teamID, Salaries.yearID \
        HAVING Batting.HBP \
        ORDER BY Salaries.teamID, Salaries.yearID DESC \
        ) AS AverageHBP \
    FROM Salaries \
    LIMIT 50;
````

## OPTIONAL UNGRADED QUESTION  

- Find the top 10 most common first names in Baseball history.


- Find the top 10 most common first names that made it into the hall of fame in baseball history.


- What have you learned?

<u>Answer:</u>

````python
# Top ten most common first names in Baseball History
%sql SELECT nameFirst, COUNT(*) AS Count \
    FROM People \
    GROUP BY nameFirst \
    ORDER BY Count DESC \
    LIMIT 10
````

````python
# Top ten most common names in the Hall of Fame
%sql SELECT nameFirst, COUNT(*) AS Count \
    FROM People \
    INNER JOIN HallOfFame ON HallOfFame.playerID = People.PlayerId \
    GROUP BY nameFirst \
    ORDER BY Count DESC \
    LIMIT 10
````
